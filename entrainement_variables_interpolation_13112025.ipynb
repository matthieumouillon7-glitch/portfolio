```python
df = pd.DataFrame({0: [1, 2, 3], 1: [4, 5, 6], 2: [7, 8, 9]})
df
```

```python
df2 = pd.DataFrame({0: [1, 2, 3], 1: [4, 5, 6], 2: [7, 8, 9]})
df2
```

```python
df=df.rename(columns={0:'Thé', 1:'Café', 2:'Biscuits'})
```

```python
df2=df2.rename(columns={0:'Chaise', 1:'Bureau', 2:'lampe'})
```

```python
merged=pd.merge(df, df2, on=df.index, how='left')
merged
```

```python
merged = pd.merge(df, df2, left_index=True, right_index=True, how='left')
print(merged)
```
