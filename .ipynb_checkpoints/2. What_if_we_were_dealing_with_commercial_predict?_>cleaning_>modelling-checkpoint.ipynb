**Retraitements**

*2 type de données numériques et non numériques, certaines à écarter d'emblée (étape1), d'autres à retraiter dans un pipeline(2), qui comportera des modèles de traitement imputer et transformer simpleimputer et RobustScaler (3) avec un columntransform (4) selon le type de données (5), puis la création d'un modèle de régression logistique linéaire (6) sur les catégories age et fare avec un predict et predict proba pour connaitre avant d'essayer le modèle. Pas de split car le jeu est étroit, ni knn.

```python
df3.columns.to_list()
```

```python
df3=df3.drop(columns=['PassengerId','Survived','Name','Ticket'])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
```

*le imputer est redondant, mais pour la logique*

*split sur les données brutes*

```python
x = pd.DataFrame(df3.drop(columns=['>100','Fare']))
  # features
y = np.array(df3['>100'].values)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
num_pipe=Pipeline([
    ('Imputer', SimpleImputer(strategy='mean')),
    ('Scaler', RobustScaler())
])
```

```python
cat_pipe=Pipeline([
    ('Imputer', SimpleImputer(strategy='most_frequent')),
    ('OHE', OneHotEncoder())
])
```

```python
preprocess = ColumnTransformer([
    ('num', num_pipe, x.select_dtypes(exclude='object').columns),
    ('cat', cat_pipe, x.select_dtypes(include='object').columns)
])
```

```python
cat_pipe
```

```python
num_pipe
```

```python
model=LogisticRegression()
```

```python
model = Pipeline([
    ('prep', preprocess),
    ('reg', model)
])
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred_pro = model.predict_proba(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(round(r2_score(y_test, y_pred),1))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, model.predict_proba(x_test)[:,1])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, model.predict(x_test))
```

*Utilisation du modèle*

```python
data_point = pd.DataFrame.from_dict({
    'Age': [60],
    'Sex': ['male'],
    'Pclass': [2],
    'Embarked': ['S'],
    'SibSp':[0],
    'Parch':[0]
})

proba_over_100 = model.predict_proba(data_point)[0][1]
proba_over_100#[0] = première (et seule) ligne du DataFrame.[1] = probabilité d’appartenir à la classe 1 (c’est-à-dire >100 euros).
```
