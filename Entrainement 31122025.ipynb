```python
#convertisseur

amount= 200
currency="EURGBP"

RATES = {
    "USDEUR": 0.85,
    "GBPEUR": 1.13,
    "CHFEUR": 0.86,
    "EURGBP": 0.885
    }

def convert(amount, currency):
    for key, value in RATES.items():
        if currency == key :
            return round(amount * value)
        
print (convert(amount, currency))
```

```python
titanic_df = pd.read_csv('https://wagon-public-datasets.s3.amazonaws.com/Machine%20Learning%20Datasets/ML_Titanic_dataset.csv')
```

```python
titanic_df['sex2']=''
```

```python
titanic_df.insert(0, 'sex2', titanic_df.pop('sex2'))
titanic_df
```

```python
def sex_num(value):
    if value == 'male':
        return 1
    else:
        return 2

# Appliquer à la colonne: 
titanic_df['sex2'] = titanic_df['Sex'].apply(sex_num)
```

```python
titanic_df['PassengerId'].nunique()
```

```python
titanic_df['Survived'].nunique()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
titanic_df = titanic_df.dropna(subset=['Age', 'Survived', 'sex2'])
X=titanic_df[['Age', 'sex2']]
y=titanic_df['Survived']
model = LogisticRegression()  # Créer l'objet
model.fit(X,y)
```

```python
# L'âge seul n'est peut-être pas assez prédictif
# Regarde le score du modèle: 
score = model.score(X, y)
print(f"Précision du modèle: {score:.2%}")
# Si c'est ~59%, le modèle prédit toujours 0 (mort)
# car 59% des gens sont morts!
```

```python
# 1. Nettoyer
titanic_df = titanic_df.dropna(subset=['Age', 'Survived', 'sex2'])

# 2. Entraîner avec Age ET sex2
X = titanic_df[['Age', 'sex2']]  # ← LES DEUX colonnes! 
y = titanic_df['Survived']

model = LogisticRegression()
model.fit(X, y)

# 3. Prédire avec Age ET sex2
new_datapoint = pd.DataFrame.from_dict({
    'Age': [10],
    'sex2': [1]
})

prediction = model.predict(new_datapoint)[0]
print(prediction)  # ✅ Fonctionne! 
```

```python
class Rectangle:
    def __init__(self, color, name, width, height): #Constructeur
        self.color = color #assigner les attributs
        self.name = name
        self.width = width
        self. height = height
    def area(self):
        return self.width * self.height
```
