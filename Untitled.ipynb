```python
a=df['children']
y=df['is_canceled']
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(17, 5))

# Subplot 1: Scatter plot
plt.subplot(2, 3, 1)
sns.scatterplot(data=df, x=a, y=y, color='green') 
plt.title('Dispersion des annulations')
plt.xlabel('children')
plt.ylabel('annulation')

# Subplot 2: Histogram
plt.subplot(2, 3, 2)
a.hist() 
plt.title('Annulation de couples avec enfants')
plt.xlabel('children')
plt.ylabel('annulation')

# Subplot 3: Regression plot
plt. subplot(2, 3, 3)
sns.regplot(data=df, x=a, y=y, color='green') 
plt.title('corrÃ©lation annulation/enfants')
plt.xlabel('children')
plt.ylabel('annulation')

# You can add more subplots (2, 3, 4), (2, 3, 5), (2, 3, 6)

plt.tight_layout()
plt.show()
```
